```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
import matplotlib
```

```python
x1 = cp.Variable(1)
x2 = cp.Variable(1)
objective_funcs = [x1 + x2, -x1 - x2, x1, cp.maximum(x1, x2), cp.power(x1, 2) + cp.multiply(9, cp.power(x2, 2))]
```

```python
obj_func_names = ['x1 + x2', '-x1 - x2', 'x1', 'max{x1, x2}', 'x1^2 + 9 * x2^2']
```

```python
for obj_func, name in zip(objective_funcs, obj_func_names):
    objective = cp.Minimize(obj_func)
    constraints = [(2 * x1) + x2 >= 1, x1 + (3 * x2) >= 1]
    prob = cp.Problem(objective, constraints)
    print('The optimal value for [{}] is:\n  |_ {}'.format(name, round(prob.solve(qcp=True), 5)))
    print('The optimal x1, x2 is:\n  |_ ({}, {})'.format(x1.value, x2.value))
    print('\n')
```

### Feasible Set

```python
plt.figure(figsize=(12, 12))

xp = np.array([-2, 20])

plt.plot(xp, 1 - (2 * xp), xp, (1 - xp) / 3) 
plt.plot(xp, [[0] for _ in xp]) 
plt.plot([[0] for _ in xp], xp)

path = matplotlib.path.Path([[0.4, 0.2], [0, 1], [1, 1], [1, 0]])            
patch = matplotlib.patches.PathPatch(path, facecolor='#C7C7ED')               
plt.gca().add_patch(patch)  

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Feasible Set')
plt.xlim(-0.05, 1)
plt.ylim(-0.05, 1)
plt.text(0.2, 0.01, "x1 = 0")
plt.text(0.01, 0.5, "x2 = 0")
plt.text(0.15, 0.8, "2 * x1 + x2 = 1")
plt.text(0.7, 0.15, "x1 + (3 * x2) = 1")
plt.show()
```
